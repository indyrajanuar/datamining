```python
import pandas as pd
import numpy as np
```

```python
#create dataset from Google Drive
dataset_url = "https://raw.githubusercontent.com/indyrajanuar/datamining/main/salaries.csv"
df=pd.read_csv(dataset_url, index_col=0)
```

```python
df
```

```python
def diss(x,y):
  a = 2
  b = 0
  for col in ['salary','salary_in_usd']:
    if df[col][x] == df[col][y]:
      b+=1
    return (a-b)/a
```

Menghitung Dissimilarity

```python
print(diss(1,1))
print(diss(1,2))
print(diss(3,3))
```
