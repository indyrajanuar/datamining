### **Transformasi data categorical menjadi numeric**

```python
# import library
import numpy as np
import pandas as pd
```

```python
#source data
dataset_url = "https://raw.githubusercontent.com/indyrajanuar/datamining/main/credit_score.xlsx"
#create dataframe
df = pd.read_excel(dataset_url)
```

```python
df
```

```python
# check information about our data
df.info()
```

```python
df.drop(columns=['kode_kontrak'])
```

```python
df[['kpr_aktif','rata_rata_overdue']]
```

```python
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()
df['kpr_aktif'] = labelencoder.fit_transform(df['kpr_aktif'])
df['rata_rata_overdue'] = labelencoder.fit_transform(df['rata_rata_overdue'])
```
